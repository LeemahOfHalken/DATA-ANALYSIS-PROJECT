## Data cleaning
* Removing irrelevant columns
 

```python
df.columns
```

* After careful study of  the correlation values in this dataset, it is of my opinion that some columns have no effect on price for simple/basic Consumers, only gamers and high-tech consumers will take note of these components before purchasing a laptop.




* Therefore, the following columns will be dropped:
- index
- secondary_storage_capacity
- gpu_brand
- is_touch_screen
- display_size
- resolution_width
- resolution_height

```python
df.drop(columns=["index","secondary_storage_capacity","is_touch_screen","gpu_brand","display_size","resolution_width","resolution_height"], inplace = True)
```

```python
avg_price= df["Price"].groupby(df["brand"]).mean().reset_index()
avg_price[0:11]
```

```python
sns.barplot(x="brand", y="Price", data= avg_price)
plt.title("Average price by brand")
plt.xlabel("brand")
plt.ylabel("Price")
plt.xticks(rotation =90);
```

* This plot indicates that apple has the most expensive brand of laptops

```python
sns.barplot(data= df, x= "OS",y="Price")
plt.ylabel("Price")
plt.xlabel("OS")
plt.title("Distribution of Price and OS")
```

* The barplot shows that the Apple Mac is the most expensive type of OS(Operating System) Software

```python
sns.histplot(data=df["processor_tier"])
plt.xlabel("brand")
plt.ylabel("Price")
plt.title("Distribution of processor_tier")
plt.xticks(rotation = 90);
```

* From the histplot, it shows that core i5 sold the most in processor tiers
